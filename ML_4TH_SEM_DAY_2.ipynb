```python
# Compute core business metrics: revenue, orders, product/city/payment/customer splits, delivery time stats
import numpy as np

# Basic KPIs
kpi_total_orders = len(df)
kpi_total_revenue = df['revenue'].sum()
kpi_aov = kpi_total_revenue / kpi_total_orders
kpi_total_items = df['quantity'].sum()
kpi_rev_per_item = kpi_total_revenue / kpi_total_items
kpi_avg_delivery = df['delivery_time_min'].mean()
kpi_p90_delivery = np.percentile(df['delivery_time_min'], 90)

print(kpi_total_orders)
print(kpi_total_revenue)
print(kpi_aov)
print(kpi_total_items)
print(kpi_rev_per_item)
print(kpi_avg_delivery)
print(kpi_p90_delivery)

# Top products by revenue and orders
prod_rev = df.groupby('product', as_index=False).agg(
    orders=('order_id','count'),
    items=('quantity','sum'),
    revenue=('revenue','sum'),
    avg_delivery=('delivery_time_min','mean')
).sort_values('revenue', ascending=False)
prod_rev['revenue_share'] = prod_rev['revenue'] / kpi_total_revenue
print(prod_rev.head(10))

# Revenue distribution by city
city_rev = df.groupby('city', as_index=False).agg(
    orders=('order_id','count'),
    revenue=('revenue','sum'),
    avg_delivery=('delivery_time_min','mean')
).sort_values('revenue', ascending=False)
city_rev['revenue_share'] = city_rev['revenue'] / kpi_total_revenue
print(city_rev.head(10))

# Payment and customer type splits
pay_rev = df.groupby('payment_method', as_index=False).agg(orders=('order_id','count'), revenue=('revenue','sum'))
pay_rev['revenue_share'] = pay_rev['revenue'] / kpi_total_revenue
pay_rev = pay_rev.sort_values('revenue', ascending=False)
print(pay_rev)

cust_rev = df.groupby('customer_type', as_index=False).agg(orders=('order_id','count'), revenue=('revenue','sum'))
cust_rev['revenue_share'] = cust_rev['revenue'] / kpi_total_revenue
cust_rev = cust_rev.sort_values('revenue', ascending=False)
print(cust_rev)

# Revenue distribution summary
rev_desc = df['revenue'].describe(percentiles=[0.25,0.5,0.75,0.9,0.95,0.99])
deliv_desc = df['delivery_time_min'].describe(percentiles=[0.25,0.5,0.75,0.9,0.95,0.99])
print(rev_desc)
print(deliv_desc)
```
